```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

final_poly_converter=PolynomialFeatures(degree=3,include_bias=False)
X_poly= final_poly_converter.fit_transform(X)

X_trai,X_tes,y_trai,y_tes=train_test_split(X_poly,y,test_size=0.3,random_state=42)

final_model=LinearRegression()
final_model.fit(X_trai,y_trai)

train_prediction= final_model.predict(X_trai)
test_prediction=final_model.predict(X_tes)

print('train r2:' , final_model.score(X_trai,y_trai))
print('test r2:' , final_model.score(X_tes,y_tes))
```

```python
X_train
```

```python
from sklearn.model_selection import cross_val_score
scores= cross_val_score(final_model,X_poly,y,cv=5)
print(scores)
scores.mean()
```

```python
from sklearn.metrics import mean_squared_error

test_RSME = np.sqrt(mean_squared_error(y_tes,test_prediction))
train_RSME = np.sqrt(mean_squared_error(y_trai,train_prediction))

print(train_RSME,test_RSME)
```

```python
from sklearn.linear_model import Lasso

final_model_lasso = Lasso(alpha=12)

final_model_lasso.fit(X_train,y_train)

test_pred=final_model_lasso.predict(X_test)
train_pred=final_model_lasso.predict(X_train)

test_r2=final_model_lasso.score(X_test,y_test)
train_r2=final_model_lasso.score(X_train,y_train)
print(train_r2)
print(test_r2)

scores= cross_val_score(final_model_lasso,X,y,cv=5)
print(scores)
scores.mean()
```

# elastic net regression


```python
from sklearn.linear_model import ElasticNet

final_model_elasticnet=ElasticNet(alpha=20,l1_ratio=1)

final_model_elasticnet.fit(X_train,y_train)

test_pred= final_model_elasticnet.predict(X_test)
train_pred=final_model_elasticnet.predict(X_train)

print('*'*50)

train_r2=final_model_elasticnet.score(X_train,y_train) # train r2
test_r2=final_model_elasticnet.score(X_test,y_test)# test r2
print('train score : ',train_r2)
print('test score : ',test_r2)

print('*'*50)

score=cross_val_score(final_model_elasticnet,X,y,cv=5)## cv
print(score)
print(score.mean())

print('*'*50)

train_rsme=np.sqrt(mean_squared_error(y_train,train_pred))# train r2
test_rsme=np.sqrt(mean_squared_error(y_test,test_pred)) # test r2
print(train_rsme)
print(test_rsme)
```
